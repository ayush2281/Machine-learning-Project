**handeling missing values**

Mean -- avg values

Mode --> most repeated value

```python
# mean value of the 'item_weight' column
big_mart_data['Item_Weight'].mean()
```

```python
# filling the missing values in 'item weight' column
big_mart_data['Item_Weight'].fillna(big_mart_data['Item_Weight'].mean(), inplace=True)
```

```python
# cheking again the null about the data
big_mart_data.isnull().sum()
```

Replacing the missing values in outlet_size with mode

```python
mode_of_outlet_size = big_mart_data.pivot_table(values='Outlet_Size', columns= 'Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode_of_outlet_size
)
```

```python
missing_values = big_mart_data['Outlet_Size'].isnull()
print(missing_values)
```

```python
# replacing the missing values with node
big_mart_data.loc[missing_values, 'Outlet_Size']= big_mart_data.loc[missing_values, 'Outlet_Type'].apply(lambda x: mode_of_outlet_size)
```

```python
# now cheking the missing values
big_mart_data.isnull().sum()
```

Data Pre-Processing

```python
big_mart_data.head()
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
big_mart_data.replace({'Item_Fat_Content': {'low fat':'Low fat','LF':'Low Fat', 'reg': 'Regular'}}, inplace = True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

label encoding

```python
encoder = LabelEncoder()
```

```python
# Check for and clean unexpected DataFrame values in 'Outlet_Size'
for i in big_mart_data.index:
    if isinstance(big_mart_data.loc[i, 'Outlet_Size'], pd.DataFrame):
        # Replace with a suitable string, e.g., 'Unknown' or most frequent value
        big_mart_data.loc[i, 'Outlet_Size'] = 'Unknown'
        # Or, investigate further to understand why a DataFrame is present

# Continue with Label Encoding after cleaning the column
big_mart_data['Item_Identifier'] = encoder.fit_transform(big_mart_data['Item_Identifier'])
big_mart_data['Item_Fat_Content'] = encoder.fit_transform(big_mart_data['Item_Fat_Content'])
big_mart_data['Item_Type'] = encoder.fit_transform(big_mart_data['Item_Type'])
big_mart_data['Outlet_Identifier'] = encoder.fit_transform(big_mart_data['Outlet_Identifier'])
big_mart_data['Outlet_Size'] = encoder.fit_transform(big_mart_data['Outlet_Size'])

big_mart_data['Outlet_Location_Type'] = encoder.fit_transform(big_mart_data['Outlet_Location_Type'])
big_mart_data['Outlet_Type'] = encoder.fit_transform(big_mart_data['Outlet_Type'])
```

Spliting the feature and target

```python
X = big_mart_data.drop(columns='Item_Outlet_Sales', axis=1)
Y = big_mart_data['Item_Outlet_Sales']
```

Spliting the data into test set and training set

```python
X_train , X_test , Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

training the Ml Model

XgBoost regressor

```python
regressor = XGBRegressor()
```

```python
regressor.fit(X_train, Y_train )
```

Evalution

```python
# predictionn on training training dat
training_data_prediction = regressor.predict(X_train)

```

```python
#  R suqaure value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
```

```python
print('R squared Value', r2_train)
```

```python
# predictionn on training training dat
test_data_prediction = regressor.predict(X_test)
```

```python
#  R suqaure value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
```
