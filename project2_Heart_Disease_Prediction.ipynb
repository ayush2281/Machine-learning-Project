Import Dependencies

```python
import numpy as np #is use to create some numpy array
import pandas as pd # creating for dataframe
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score # evaluate our model
```

Data collection and Processing

```python
# loading the csv data to a Pandas Dataframe
heart_data = pd.read_csv('/content/heart_disease_data3.csv')
```

```python
# statistical measure about the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data['target'].value_counts()
```


*   
   1 --> **defective** **Heart**
*  0 --> **Healthy** **Heart**






Splitting the Feature and Target

```python
X = heart_data.drop(columns='target',axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Spilitting the data into training data and test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

X_train the features contains all the training data,

X_test is contain the feature  all the test  data

Y_train is contain the target of all these features present in the x train

Y_test  target for X_test

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistics Regression

```python
model = LogisticRegression()
```

```python
# training the Logistics Regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuarcy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
# accuarcy on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data: ", test_data_accuracy)
```

Building Predictive Function

```python
input_data= (41,0,1,130,204,0,0,172,0,1.4,2,0,2)

# change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting only one instance

input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] --0):
  print("the person does not have Heart disease")
else:
  print("the person have heart Disease")
```
