```python
X = mall_dataset.iloc[:,[3,4]].values #iloc = licate the column in tha dataset
```

```python
print(X)
```

choosing the number of clusters

WCSS - Within cluster sum of squred

```python
# finding the wcss value for different number of cluster
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
# plot the elbow graph
sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of cluster = 5

training the k-means clustering model

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

# return a label fro each data point based on thier cluster
Y = kmeans.fit_predict(X)
print(Y)
```

5 clusters - 0,1,2,,3,4

Visulaize all clusters

```python
# plotung the all the clusterts and their Centroids

plt.figure(figsize=(10,10))
plt.scatter(X[Y==0,0], X[Y==0,1],s=50, c='green', label = 'Cluster1')
plt.scatter(X[Y==1,0], X[Y==1,1],s=50, c='yellow', label = 'Cluster2')
plt.scatter(X[Y==2,0], X[Y==2,1],s=50, c='red', label = 'Cluster3')
plt.scatter(X[Y==3,0], X[Y==3,1],s=50, c='orange', label = 'Cluster4')
plt.scatter(X[Y==4,0], X[Y==4,1],s=50, c='blue', label = 'Cluster5')

# plot the centroid
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='black', label = 'Centroids')

plt.title('Cluster of costumer')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()





```
