```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and Processing

```python
#loading dataset from sklearn
breast_cancer_dataset  = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 row  of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to thhe data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# printing the last 5 rows  of the dataframe
data_frame.tail()
```

```python
# number of the rows and columnn in the dataset
data_frame.shape
```

```python
# getting some information about the data fram
data_frame.info()
```

```python
# checkinhg the missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Variable
data_frame['label'].value_counts()
```

0 --> malignant

1 --> benign

```python
data_frame.groupby('label').mean()
```

Seprating the Features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Spliting the data into training data and test data

```python
X_train, X_test , Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# traing the logistic regression model
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy score on the training
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy Score on the Training data =', training_data_accuracy)
```

```python
# accuracy score on the training
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy Score on the Test data =', test_data_accuracy)
```

**Building a Predictive System**

```python
input_data = (12.45,15.7,82.57,477.1,0.1278,0.17,0.1578,0.08089,0.2087,0.07613,0.3345,0.8902,2.217,27.19,0.00751,0.03345,0.03672,0.01137,0.02165,0.005082,15.47,23.75,103.4,741.6,0.1791,0.5249,0.5355,0.1741,0.3985,0.1244

)
# change the input data to anumpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
